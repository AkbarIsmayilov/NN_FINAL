```python
img = io.imread('figure1.png')
# getting image width and height
width = img.shape[0]
height = img.shape[1]
sample_range  =  range (10 ,10000, 100)
print(width, height)

img= img[:,:,1] 
  
# using deepcopy for deepcopy  
img1 = copy.deepcopy(img) 
# img1 =  img
# img1 = img1[:,:,1] 
# print(img1.shape  ,img.shape    )
# img1 = img 
```

```python

def getArea(sample_size) : 
  cnt = 0

  for i in range(sample_size):

    x = randint(0,width-1)
    y = randint(0,height-1)
    # img1[x][y] =  
    
    if img[x][y] != 255:
      cnt += 1
      img1[x][y] = 255
    else : 
      img1[x][y] = 0
  # print(cnt)
  return (cnt / sample_size) * (width * height)
```

```python
area = getArea(10000)

print(area)
io.imshow(img1) 
```

```python

stats  = list()

for i in sample_range : 
  # img1 = copy.deepcopy(img) 

  area = getArea(i)
  stats.append(area)
```
